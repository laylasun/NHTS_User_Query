```python
# import files 2016&2017
veh16 = pd.read_csv(f_veh16)
veh17 = pd.read_csv(f_veh17)
```

```python
# 4/2016-4/2017; no states 43 and 52
veh16_trim = veh16[(veh16['MONTH'] >= 4) & (veh16['STATE'] != 43) & (veh16['STATE'] != 52)]
veh17_trim = veh17[(veh17['MONTH'] <= 4) & (veh17['STATE'] != 43) & (veh17['STATE'] != 52)]
```

```python
# recode for vehicle type
def veh_type(row):
    r = row['BODY_TYP']
    if r >= 1 and r <= 10:
        return '1-Auto'
    elif r in [14,15,16,17,19]:
        return '3-SUV'
    elif r in [20,21,22,28,29]:
        return '2-Van'
    elif r in [30,31,32,34,39,49]:
        return '4-Pickup Truck'
    elif r in [80,81,82,83,84,85,88,89]:
        return '7-Motorcycle'
    else:
        return '9-Other'
```

```python
# combine veh16_trim and veh2017_trim
veh1617_trim = veh16_trim.append(veh17_trim)
veh1617_trim['VEH_TYPE'] = veh1617_trim.apply(veh_type, axis = 1)
len(veh1617_trim)
```

```python
# export to file
f_veh_res = r'E:\GIS_Data\Layla2019_NHTS_backup\internal_task\jan2021_traffic safety data report\fars data\output tables\veh_57413_vehtyp.csv'
veh1617_trim.to_csv(f_veh_res)
```
